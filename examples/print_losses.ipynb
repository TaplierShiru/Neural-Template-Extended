```python
base_path = '/path/to/res/train_neural_template/class_run_1/'
test_line = 'Test Loss for epoch'

with open(os.path.join(base_path, '2023-06-28_00-04-53_IM-Net-Training-experiment-home-AutoEncoder-3DCNN-Flow/print.log'), 'r') as rf:
    lines = rf.readlines()

loss_stage_con_0 = []
    
for line in lines:
    if test_line not in line:
        continue
    loss_stage_con_0.append(float(line.split(':')[-1].strip()))
```

```python
with open(os.path.join(base_path, '2023-07-01_11-31-15_IM-Net-Training-experiment-home-AutoEncoder-3DCNN-Flow/print.log'), 'r') as rf:
    lines = rf.readlines()

phase_line = 'Start Phase'
phase = 0

loss_stage_con_1 = []
loss_stage_con_2 = []

for line in lines:
    if phase_line in line:
        phase = int(line.strip()[-1])
    if test_line not in line:
        continue
    value = float(line.split(':')[-1].strip())
    if phase == 0:
        loss_stage_con_0.append(value)
    elif phase == 1:
        loss_stage_con_1.append(value)
    elif phase == 2:
        loss_stage_con_2.append(value)
```

```python
fig = plt.figure(figsize=(12, 6))

ax = fig.add_subplot(131)
plt.plot(loss_stage_con_0)
plt.title('con 0')

ax = fig.add_subplot(132)
plt.plot(loss_stage_con_1)
plt.title('con 1')

ax = fig.add_subplot(133)
plt.plot(loss_stage_con_2)
plt.title('con 2')
```
