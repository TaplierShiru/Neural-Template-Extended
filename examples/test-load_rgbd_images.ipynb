```python
from PIL import Image
# !pip install pillow-heif
from pillow_heif import register_heif_opener

register_heif_opener()
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = './rgbd_imgs/divan/IMG_0002.HEIC'
depth_path = img_path.replace('.HEIC', '_depth.TIF')

depth = Image.open(depth_path)
depth = depth.transpose(Image.Transpose.ROTATE_270)
depth_np = np.array(depth)

img = Image.open(img_path)
img = img.resize(depth.size)
img_np = np.array(img)
```

```python
depth_np.shape, depth_np.mean(), depth_np.max(), depth_np.min()
```

```python
fig = plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.imshow(img_np)

plt.subplot(122)
plt.imshow(depth_np, cmap='hot', interpolation='nearest')
```
